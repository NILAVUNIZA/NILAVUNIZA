```python
from netCDF4 import Dataset
import pandas as pd
import xarray as xr

# Open the .nc file using xarray
ds = xr.open_dataset('/content/GODAS.input.36mn.1980_2015.nc')

# Convert the dataset to a pandas DataFrame
df = ds.to_dataframe()

# Display the first few rows of the DataFrame
display(df.head())
```

```python
# Save the DataFrame to a CSV file
df.to_csv('godas_data.csv')

# Display the first few rows of the CSV file to verify
csv_df = pd.read_csv('godas_data.csv')
display(csv_df.head())
```

```python
from netCDF4 import Dataset
import pandas as pd
import xarray as xr

# Open the .nc file using xarray
ds = xr.open_dataset('/content/GODAS.input.type.NDJ.1980_2015.nc')

# Convert the dataset to a pandas DataFrame
df = ds.to_dataframe()

# Display the first few rows of the DataFrame
display(df.head())
```

```python
from netCDF4 import Dataset
import pandas as pd
import xarray as xr

# Open the .nc file using xarray, specifying decode_times with CFDatetimeCoder
time_coder = xr.coders.CFDatetimeCoder(use_cftime=True)
ds = xr.open_dataset('/content/GODAS.input.type.NDJ.1980_2015.nc', decode_times=time_coder)

# Convert the dataset to a pandas DataFrame
df = ds.to_dataframe()

# Display the first few rows of the DataFrame
display(df.head())
```

```python
# Save the DataFrame to a CSV file
df.to_csv('godas_data_ndj.csv')

# Display the first few rows of the CSV file to verify
csv_df_ndj = pd.read_csv('godas_data_ndj.csv')
display(csv_df_ndj.head())
```
